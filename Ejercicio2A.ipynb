```python
#Importar librerias de collections
from collections import Counter, OrderedDict

#Lista de datos iniciales
compras=["Luis","Ana","Luis","Carlos","Marta","Ana","Sofía","Elena","Luis","Carlos"]
registrados=["Ana","Carlos","Marta","Elena"]

#Filtrar clientes nuevos
#Usuarios no registrados
usuarios_noregistrados=set(compras)-set(registrados)
print(set(compras).difference(set(registrados)))
```

```python
#Eliminar duplicados y mantener el orden
#print(set(compras))
#crear lista
clientes_unicos=list(OrderedDict.fromkeys(compras))
```

```python
#Contar cuántas veces se repite cada nombre
conteo_clientes=Counter(compras)
#print(conteo_clientes)
for usuario,veces in conteo_clientes.items():
    print(f"{usuario} aparece {veces} en la lista de compras")
```

```python
#Crear un resumen personalizado
# dict comprehension | variables: usuario, veces | donde veces>1 (ha comprado mas de una vez)
cliente_frecuente={usuario:f"ha comprado {veces} veces" for usuario,veces in conteo_clientes.items() if veces>1}
print(cliente_frecuente)
```

```python
#Formato final
print(f"Los usuarios no registrados son: {usuarios_noregistrados}")
print(f"\nLa lista de clientes unicos es la siguiente:")
for i in clientes_unicos:
    print(i)

print("\nClientes frecuentes:")
for i in cliente_frecuente:
    print(f"{i} -> {cliente_frecuente[i]}")
```
